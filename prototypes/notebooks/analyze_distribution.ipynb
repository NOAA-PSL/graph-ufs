## Atmosphere

```python
# 2D variable
varname_2d = "spfh2m"
transformation = "sqrt"   # e.g., log, sqrt
t0 = 0; tend = 10000
isdensity = True
```

```python
# slice the variable
var_sliced = atm[varname_2d].isel(time=slice(t0, tend)).load()
```

```python
# Transform and plot if specified, else simply plot
if transformation:
    try:
        func = getattr(np, transformation)
    except AttributeError:
        try:
            func = globals()[transformation]
        except KeyError:
            raise KeyError(f"{transformation} neither found in numpy or in the global scope")
    except Exception as e:
        raise Error(f"unknown exception {e} occured")
        
    transformed_var = func(var_sliced)
    plt.hist(transformed_var.values.flatten(), bins=25, color='skyblue', 
             edgecolor='white', density=isdensity)
    title = f"Histogram - {transformation}({varname_2d})"
    figname = f"distribution_{transformation}({varname_2d})_nt-{int(tend-t0)}.png"
else:
    plt.hist(var_sliced.values.flatten(), bins=25, color='skyblue', 
             edgecolor='white', density=isdensity)
    title = f"Histogram - {varname_2d}"
    figname = f"distribution_{varname_2d}_nt-{int(tend-t0)}.png"

plt.title(title)
plt.xlabel('Values')
plt.ylabel('Density')
plt.grid(alpha=0.2)

#plt.savefig(os.path.join(figdir, figname), dpi=300)
plt.show()
```

```python
# 3D variable
varname_2d = "spfh"
presslevel = 676 # in hPa
transformation = None   # e.g., "log", "sqrt"
t0 = 0; tend = 10000
isdensity = True
```

```python
# slice the variable
var_sliced = atm[varname_2d].isel(time=slice(t0, tend)).sel(pfull=presslevel, method="nearest").load()
```

```python
# Transform and plot if specified, else simply plot
if transformation:
    func = getattr(np, transformation)
    transformed_var = func(var_sliced)
    plt.hist(transformed_var.values.flatten(), bins=25, color='skyblue', 
             edgecolor='white', range=(-12,-2), density=isdensity)
    title = f"Histogram - {transformation}({varname_2d})"
    figname = f"distribution_{transformation}({varname_2d})_nt-{int(tend-t0)}_plevel-{presslevel}.png"
else:
    plt.hist(var_sliced.values.flatten(), bins=25, color='skyblue', 
             edgecolor='white', density=isdensity)
    title = f"Histogram - {varname_2d}"
    figname = f"distribution_{varname_2d}_nt-{int(tend-t0)}_plevel-{presslevel}.png"

plt.title(title)
plt.xlabel('Values')
plt.ylabel('Density')
plt.grid(alpha=0.2)

plt.savefig(os.path.join(figdir, figname), dpi=300)
plt.show()
```
